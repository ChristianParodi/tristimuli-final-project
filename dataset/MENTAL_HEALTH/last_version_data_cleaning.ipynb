```python
import pandas as pd

deaths = pd.read_csv('./clean/estat_hlth_cd_aro.csv')
deaths
```

```python
deaths_melted = deaths.melt(id_vars=['country', 'cause', 'age', 'sex'], 
							value_vars=['2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'], 
							var_name='year', 
							value_name='deaths')
deaths_melted
```

```python
percentages = []

for index, row in deaths_melted.iterrows():
    if index%1000 == 0:
        print(index)
    
    total_deaths = deaths_melted[
		(deaths_melted['country'] == row['country']) & 
		(deaths_melted['cause'] == 'Total') & 
		(deaths_melted['age'] == 'Total') & 
		(deaths_melted['sex'] == 'Total') & 
		(deaths_melted['year'] == row['year'])
	]['deaths'].values
    
    if total_deaths.size > 0:
        percentage = (row['deaths'] / total_deaths[0]) * 100
    else:
        percentage = 0
        
    percentages.append(percentage)

deaths_melted['percentage'] = percentages
deaths_melted
```

```python
ages = [("65 years or over", "65"), ("From 15 to 29 years", "15-29"), ("From 30 to 44 years", "30-44"), ("From 45 to 64 years", "45-64"), ("Less than 15 years", "15")]
for fr, to in ages:
    deaths_melted.loc[deaths_melted['age'] == fr, 'age'] = to

deaths_melted.age.unique()
```

```python
deaths_melted.to_csv('./clean/melted_estat_hlth_cd_aro.csv', index=False)
```
