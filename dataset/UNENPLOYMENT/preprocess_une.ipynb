```python
import pandas as pd
import numpy as np
import glob
```

```python
df = pd.read_csv('estat_une_rt_m_filtered.csv')
df.columns = df.columns.str.strip()
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df.head()
```

```python
path = '../ESTAT_CODELISTS/*.tsv'
files = glob.glob(path)
dataframes = {file.split('/')[-1].replace('.tsv', ''): pd.read_csv(
    file, sep='\t', on_bad_lines='skip') for file in files}
```

```python
for col in ["freq", "unit", "sex", "age", "s_adj", "geo"]:
    curr_codelist = dataframes[f"ESTAT_{col.upper()}"]
    df[col] = df[col].map(curr_codelist.set_index('CODE')['Label'])

df.rename(columns={'geo': 'country'}, inplace=True)
df.head()
```

```python
df.replace(": z", np.nan, inplace=True)
df.replace(":", np.nan, inplace=True)
```

```python
year_cols = [col for col in df.columns if col.isdigit()]

for col in year_cols:
    df[col] = df[col].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
df.head()
```

```python
for col in df.columns:
    if not col.isdigit() and df[col].nunique() == 1:
        df.drop(columns=[col], inplace=True)
```

```python
df.head()
```

```python
# Check the distinct values in s_adj
print("Distinct values in s_adj:", df["s_adj"].unique())

# Example approach: keep only rows where s_adj is the first distinct value
val_to_keep = df["s_adj"].dropna().unique()[0]  # keep seasonally adjusted
df = df[df["s_adj"] == val_to_keep]
df.drop(columns=["s_adj"])
```

```python
df['country'] = df["country"].replace('TÃ¼rkiye', 'Turkey')
```

```python
df = df.loc[df["country"] != "European Union - 28 countries (2013-2020)"]
df.loc[:, 'country'] = df["country"].replace('European Union - 27 countries (from 2020)', 'European Union')
```

```python
month_cols = [c for c in df.columns if '-' in c and c[:4].isdigit()]
df[month_cols] = df[month_cols].apply(pd.to_numeric, errors='coerce')
df['mean_unenrollments'] = df[month_cols].mean(axis=1)

country_with_max = df.groupby('country')['mean_unenrollments'].mean().idxmax()
print("Country with maximum mean unenrollments:", country_with_max)
```

```python
df.to_csv("./clean/estat_une_rt_m_filtered.csv", index=False)
```
