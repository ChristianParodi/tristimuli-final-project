```python
import pandas as pd
```

```python
country_codes = pd.read_json('../countries_ISO2_ISO3.json')
```

```python
df = pd.read_csv('lockdowns_2022-08-25.csv')
df.columns = df.columns.str.strip()
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df.head()
```

```python
var_dict = pd.read_csv("variable_dictionary.csv")
var_dict.head()
```

```python
df["description"] = df["measure"].map(
    var_dict.set_index('measure')['description'])

df = df[['country', 'measure', 'description', 'date_start', 'date_end']]
```

```python
df.head()
```

```python
important_measures = [
    "StayHomeOrder",
    "StayHomeGen",
    "SocialCircle",
    "MasksMandatoryAllSpaces",
    "HotelsAccommodation",
    "ClosDaycare",
    "ClosPrim",
    "ClosSec",
    "ClosHigh",
    "MassGatherAll",
    "ClosPubAny",
    "Teleworking",
    "WorkplaceClosures",
    "ClosureOfPublicTransport",
    "QuarantineForInternationalTravellers",
]
```

```python
df = df[df['measure'].isin(important_measures)].reset_index(drop = True)
df.head()
```

```python
df = country_codes.merge(df, left_on = 'country', right_on = 'country', how = 'inner')
df
```

```python
df.to_csv("./clean/lockdonws_2022-08-25.csv", index=False)
```
