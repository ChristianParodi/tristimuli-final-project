```python
import pandas as pd

europe = pd.read_csv('covid.csv')
europe.info()
```

```python

people_vacc_empty = (europe['people_vaccinated'].isnull()) | (
    europe['people_vaccinated'] == 0)
new_vacc_not_empty = (europe['new_vaccinations'] > 0)

peop_emp = europe[people_vacc_empty][['people_vaccinated', 'new_vaccinations']]
```

```python
peop_emp[peop_emp['new_vaccinations'] > 0]
```

```python
europe[(europe['country'] == "Albania") & (europe['year'] == 2021)
       & (europe['month'] == 5)]['people_vaccinated']
```

```python
for i in range(1, len(europe)):
    if europe.iloc[i]['country'] == europe.iloc[i-1]['country']:
        if (europe.iloc[i]['people_vaccinated'] == 0) or (pd.isnull(europe.iloc[i]['people_vaccinated'])):
            europe.loc[i, 'people_vaccinated'] = europe.iloc[i -
                                                             1]['people_vaccinated'] + europe.iloc[i]['new_vaccinations']
```
