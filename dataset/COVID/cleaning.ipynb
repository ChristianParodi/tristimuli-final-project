```python
import pandas as pd

data = pd.read_csv('covid.csv')
col_desc = pd.read_csv('columns_description.csv')
```

```python
data.head()
```

```python
europe = data[data['continent']=='Europe'].reset_index(drop=True)
```

```python
europe
```

## 31/12/2024

```python
idx = (europe['total_cases'].isnull()) & (europe['year'] == 2024) & (europe['month']==12)
europe[idx]
```

```python
true_idx_values = idx[idx].index.values
for i in true_idx_values:
    for col in europe.columns:
        if "new" in col:
            europe.loc[i, col] = 0
        else:
            europe.loc[i, col] = europe.loc[i-1, col]
```

```python
europe[idx]
```

## Pre marzo 2020

```python
idx = (europe['total_cases'].isnull())
europe.loc[idx, 'total_cases']=0.0
europe[idx]
```

```python
for col in europe.columns:
    if "new" in col or "total" in col or "cases" in col:
        if europe.loc[idx, col].isnull().values.any():
        	europe.loc[idx, col] = 0.0
        
europe[idx]
```

```python
cols_to_zero = ['new_cases', 'new_cases_smoothed', 'total_cases_per_million',
       'new_cases_per_million', 'new_cases_smoothed_per_million',
       'total_deaths', 'new_deaths', 'new_deaths_smoothed',
       'total_deaths_per_million', 'new_deaths_per_million',
       'new_deaths_smoothed_per_million', 'hosp_patients',
       'hosp_patients_per_million', 'weekly_hosp_admissions',
       'weekly_hosp_admissions_per_million', 'icu_patients',
       'icu_patients_per_million', 'weekly_icu_admissions',
       'weekly_icu_admissions_per_million',
       'reproduction_rate', 'total_tests', 'new_tests',
       'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
       'positive_rate', 'tests_per_case', 'total_vaccinations',
       'people_vaccinated', 'people_fully_vaccinated', 'total_boosters',
       'new_vaccinations', 'new_vaccinations_smoothed',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred',
       'new_vaccinations_smoothed_per_million',
       'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred']
```

```python
idx = europe['total_cases']==0
idx_to_zero = idx[idx].index.values

for i in idx_to_zero:
    for col in cols_to_zero:
        if pd.isna(europe.loc[i, col]):
            europe.loc[i, col] = 0
```
